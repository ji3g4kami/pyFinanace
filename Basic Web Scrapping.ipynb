# 用 requests.get 爬取資料

```python
import requests
url = 'http://www.tse.com.tw/ch/trading/fund/T86/T86.php'
res = requests.get(url)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
```

# 用 requests.post 爬取需傳送額外資訊的資料

```python
import requests
from bs4 import BeautifulSoup
url = 'http://www.tse.com.tw/ch/trading/fund/T86/T86.php'
form_data = {
    'qdate':'106/04/14',
    'select2':'24',
    'sorting':'by_issue'
}
res = requests.post(url, form_data)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
table = soup.select('#tbl-sortable')[0]
type(table)
```
